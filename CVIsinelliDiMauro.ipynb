# Importazione librerie

```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Funzione per caricare le immagini da una cartella
def load_images_from_folder(folder_path, img_size=(128, 128)):
    images = []
    labels = []
    class_names = os.listdir(folder_path)
    
    for label, class_name in enumerate(class_names):
        class_path = os.path.join(folder_path, class_name)
        if os.path.isdir(class_path):
            for filename in os.listdir(class_path):
                if filename.endswith(".jpg") or filename.endswith(".png"):
                    img_path = os.path.join(class_path, filename)
                    img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
                    img = cv2.resize(img, img_size)
                    images.append(img)
                    labels.append(label)
    return np.array(images), np.array(labels), class_names

# Percorso delle immagini (cambia questo con il percorso del tuo dataset)
dataset_path = "/Users/giulia/Documents/GitHub/CV_DiMauro-Isinelli/ISIC_dataset"

# Carica le immagini
images, labels, class_names = load_images_from_folder(dataset_path)

# Stampa i dettagli sul dataset
print(f"Numero di immagini: {len(images)}")
print(f"Numero di classi: {len(class_names)}")
print(f"Nomi delle classi: {class_names}")

# Funzione per visualizzare alcune immagini
def plot_sample_images(images, labels, class_names, num_samples=5):
    plt.figure(figsize=(10, 5))
    for i in range(num_samples):
        plt.subplot(1, num_samples, i + 1)
        plt.imshow(images[i], cmap='gray')
        plt.title(class_names[labels[i]])
        plt.axis('off')
    plt.tight_layout()
    plt.show()

# Visualizza un campione di immagini
plot_sample_images(images, labels, class_names)
```
